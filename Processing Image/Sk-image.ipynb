```python
from skimage import data, color, io
from matplotlib import pyplot as plt
```

```python
rocket=data.rocket()
gray_scale_rocket=color.rgb2gray(rocket)
#Figure 1
f1=plt.figure(1)
io.imshow(rocket)
#Figure 2
f2=plt.figure(2)
io.imshow(gray_scale_rocket)
plt.show()
```

```python
from skimage.transform import resize
chelsea=data.chelsea()
resized_image=resize(chelsea,(70,70), order=0, preserve_range=True, 
                     anti_aliasing=False).astype('uint8')
io.imshow(resized_image)
io.show()
```

```python
from skimage.transform import rotate
```

```python
rotate_image=rotate(chelsea, angle=80, resize=True)
io.imshow(rotate_image)
io.show()
```

```python
from skimage.filters import sobel
image=data.coins()
io.imshow(image)
io.show()

image_sobel=sobel(image)
io.imshow(image_sobel)
io.show()
```

### Segmentation

```python
from skimage import io, segmentation
import matplotlib.pyplot as plt

n_segments=6
fig_width=2.5*n_segments
img=data.coins()
segments=segmentation.slic(img, n_segments=n_segments)

fig, ax=plt.subplots(1, n_segments)
fig.set_figwidth(fig_width)

for index in np.unique(segments):
    segment=img.copy()
    segment[segments != index]=0
    ax[index].imshow(segment)
    ax[index].set(title=f'Segment {index}')
    ax[index].set_axis_off()
```
