#### ROTATING Image

```python
(w,h)=img.shape[:2]
center=(w//2,h//2)
```

```python
M=cv.getRotationMatrix2D(center,-45,1.0)
rotated=cv.warpAffine(img, M,(w,h))
cv.imshow('OpenCV Rotation',rotated)
cv.waitKey(0)
cv.destroyAllWindows()
print(M)
```

```python
face_cascade=cv.CascadeClassifier('C:\\Users\\hacker\\Anaconda3\\Lib\\site-packages\\cv2\\data\\haarcascade_frontalface_default.xml')
eyes_cascade=cv.CascadeClassifier('C:\\Users\\hacker\\Anaconda3\\Lib\\site-packages\\cv2\\data\\haarcascade_eye.xml')
```

```python
img_face=cv.imread('C:\\Users\\hacker\\Desktop\\pictures\\image\\Camera\\IMG_20170522_193918224.jpg',1)
plt.imshow(img_face)
```

```python
gray=cv.cvtColor(img_face, cv.COLOR_BGR2GRAY)
faces=face_cascade.detectMultiScale(gray,1.3, 5)
for(x,y,w,h) in faces:
    img_face=cv.rectangle(img_face,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray=gray[y:y+h, x:x+h]
    roi_color=img_face[y:y+h,x:x+w]
    eyes=eyes_cascade.detectMultiScale(roi_gray)
    for(ex,ey,ew,eh) in eyes:
        cv.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)

cv.imshow('Image Face eyes Detecting',img_face)
cv.waitKey(0)
cv.destroyAllWindows()
```
